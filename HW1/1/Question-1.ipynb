```python

import  numpy as np 

def threshold (a):
    return 1 if a>=2 else 0
def XOR(A0,B0):
    w1=[2,-1]
    w3=[2,2]
    z1=threshold(np.sum(np.dot (w1,[A0,B0])))
    z2=threshold(np.sum(np.dot (w1,[B0,A0])))
    Y=threshold(np.sum(np.dot (w3,[z1,z2])))
    return Y

def AND(A0,B0):
    w1=[1,1]
    Y=threshold(np.sum(np.dot (w1,[A0,B0])))
    return Y

def OR(A0,B0):
    w1=[2,2]
    Y=threshold(np.sum(np.dot (w1,[A0,B0])))
    return Y

def SUM_1(A1,A0,B1,B0):
    return XOR(A0,B0)

def SUM_2(A1,A0,B1,B0):
    z1=AND(A0,B0)
    z2=XOR(A1,B1)
    return XOR(z1,z2)

def CARRYOUT(A1,A0,B1,B0):
    z1=AND(A1,B1)
    z2=AND(A0,B0)
    z3=XOR(A1,B1)
    z4=AND(z3,z2)
    return OR(z4,z1)

def FullAdder(A1,A0,B1,B0):
    return CARRYOUT(A1,A0,B1,B0),SUM_2(A1,A0,B1,B0),SUM_1(A1,A0,B1,B0)

for i in range(16):
  number = format(i,'004b')
  x = FullAdder(int(number[0]),int(number[1]),int(number[2]),int(number[3]))
  print(number)
  print(x)

```
